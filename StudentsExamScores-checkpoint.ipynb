```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
df.head()
```

# <span style="color: red;">Drop Unnamed Column</span>

```python
df = df.drop("Unnamed: 0" , axis = 1)
df.head()
```

# <span style="color: red;">Gender Distribution</span>

```python
plt.figure(figsize = (4 , 3))
sns.countplot(data = df , x = "Gender" , hue = "Gender")
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize = (4 , 5))
ax = sns.countplot(data = df , x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

## <span style="color: red;">from the graph we have analysed:</span>
### the number of males in the data is less than number of females

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean' , "ReadingScore":'mean' , "WritingScore":'mean'})
gb.head()
```

```python
plt.figure(figsize = (6 , 4))
sns.heatmap(gb , annot = True)
plt.title("Parent's Education and Student's Score")
plt.show()
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean' , "ReadingScore":'mean' , "WritingScore":'mean'})
gb1.head()
```

```python
plt.figure(figsize = (8 , 5))
sns.heatmap(gb1 , annot = True)
plt.title("Parent's Marital Status and Student's Score")
plt.show()
```

# <span style="color: red">Distribution of Ethnic Groups</span>

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupA.head(10)
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A" , "group B" , "group C" , "group D" , "group E"]
groupList = [groupA["EthnicGroup"] , groupB["EthnicGroup"] , groupC["EthnicGroup"] , groupD["EthnicGroup"] , groupE["EthnicGroup"]]
plt.pie(groupList , labels = l , autopct = "%1.2f%%")

print(groupList)

plt.title("Distribution of Ethnic Groups")
```
